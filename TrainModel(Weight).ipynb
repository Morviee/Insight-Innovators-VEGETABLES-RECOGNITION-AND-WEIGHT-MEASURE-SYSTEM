```python
from google.colab import drive
import tensorflow as tf
import matplotlib.pyplot as plt
import pandas as pd

drive.mount('/content/drive')

# Load training set for classification
training_set_classification = tf.keras.utils.image_dataset_from_directory(
    '/content/drive/MyDrive/Vegetable Images/train',
    labels="inferred",
    label_mode="categorical",
    class_names=None,
    color_mode="rgb",
    batch_size=40,
    image_size=(64, 64),
    shuffle=True,
    seed=None,
    validation_split=None,
    subset=None,
    interpolation="bilinear",
    follow_links=False,
    crop_to_aspect_ratio=False
)

# Load validation set for classification
validation_set_classification = tf.keras.utils.image_dataset_from_directory(
    '/content/drive/MyDrive/Vegetable Images/validation',
    labels="inferred",
    label_mode="categorical",
    class_names=None,
    color_mode="rgb",
    batch_size=40,
    image_size=(64, 64),
    shuffle=True,
    seed=None,
    validation_split=None,
    subset=None,
    interpolation="bilinear",
    follow_links=False,
    crop_to_aspect_ratio=False
)

# Load training set for regression
training_set_regression = tf.keras.utils.image_dataset_from_directory(
    '/content/drive/MyDrive/Vegetable Images/train',
    labels="inferred",
    label_mode="int",  # Assuming you have a numerical label for weight
    class_names=None,
    color_mode="rgb",
    batch_size=40,
    image_size=(64, 64),
    shuffle=True,
    seed=None,
    validation_split=None,
    subset=None,
    interpolation="bilinear",
    follow_links=False,
    crop_to_aspect_ratio=False
)

# Load validation set for regression
validation_set_regression = tf.keras.utils.image_dataset_from_directory(
    '/content/drive/MyDrive/Vegetable Images/validation',
    labels="inferred",
    label_mode="int",  # Assuming you have a numerical label for weight
    class_names=None,
    color_mode="rgb",
    batch_size=40,
    image_size=(64, 64),
    shuffle=True,
    seed=None,
    validation_split=None,
    subset=None,
    interpolation="bilinear",
    follow_links=False,
    crop_to_aspect_ratio=False
)

# Create and compile the classification model
classification_model = tf.keras.models.Sequential()
classification_model.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, padding='same', activation='relu', input_shape=[64, 64, 3]))
classification_model.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu'))
classification_model.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
classification_model.add(tf.keras.layers.Dropout(0.25))
classification_model.add(tf.keras.layers.Conv2D(filters=64, kernel_size=3, padding='same', activation='relu'))
classification_model.add(tf.keras.layers.Conv2D(filters=64, kernel_size=3, activation='relu'))
classification_model.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
classification_model.add(tf.keras.layers.Dropout(0.25))
classification_model.add(tf.keras.layers.Flatten())
classification_model.add(tf.keras.layers.Dense(units=512, activation='relu'))
classification_model.add(tf.keras.layers.Dense(units=256, activation='relu'))
classification_model.add(tf.keras.layers.Dropout(0.5))
classification_model.add(tf.keras.layers.Dense(units=15, activation='softmax'))  # Assuming 15 classes for vegetables
classification_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
classification_model.summary()

# Create and compile the regression model
regression_model = tf.keras.models.Sequential()
regression_model.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, padding='same', activation='relu', input_shape=[64, 64, 3]))
regression_model.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu'))
regression_model.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
regression_model.add(tf.keras.layers.Dropout(0.25))
regression_model.add(tf.keras.layers.Conv2D(filters=64, kernel_size=3, padding='same', activation='relu'))
regression_model.add(tf.keras.layers.Conv2D(filters=64, kernel_size=3, activation='relu'))
regression_model.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
regression_model.add(tf.keras.layers.Dropout(0.25))
regression_model.add(tf.keras.layers.Flatten())
regression_model.add(tf.keras.layers.Dense(units=512, activation='relu'))
regression_model.add(tf.keras.layers.Dense(units=256, activation='relu'))
regression_model.add(tf.keras.layers.Dropout(0.5))
regression_model.add(tf.keras.layers.Dense(units=1, activation='linear'))  # Output layer for regression
regression_model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mean_absolute_error'])
regression_model.summary()

# Train the classification model
classification_history = classification_model.fit(x=training_set_classification, validation_data=validation_set_classification, epochs=32)

# Train the regression model
regression_history = regression_model.fit(x=training_set_regression, y=train_weights_data['weight'], validation_data=(validation_set_regression, validation_weights_data['weight']), epochs=32)
```
